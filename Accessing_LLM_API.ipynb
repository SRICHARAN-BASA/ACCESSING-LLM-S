**Accessing Groq LLM using API in python:**

```python
import groq
import os
from groq import Groq
```

```python
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")

client = Groq()

completion = client.chat.completions.create(

    model = 'llama-3.3-70b-versatile',
    messages=[
        {
            "role": "user",
            "content": "What is python and explain it in simple terms"
        }
    ]
)

print(completion.choices[0].message.content)
```

**Creating chat completion with passing chat history**

```python
from groq import Groq
client = Groq()
#create a chathistory with system, user and assistant messages
chathistory = [
    {
        "role": "system",
        "content": "You are a helpful assistant that explains things in simple terms."
    },
    {
        "role": "user",
        "content": "What is python and explain it in simple terms"
    },
    {
        "role": "assistant",
        "content": "Python is a popular programming language that is easy to learn and use. It is known for its simplicity and readability, making it a great choice for beginners. Python can be used for a wide range of applications, including web development, data analysis, artificial intelligence, and more."
    },
    {
        "role": "user",
        "content": "What are some popular libraries in python?"
    }
]

#pass the chathistory to the model and get a response
completion = client.chat.completions.create(
    model = 'llama-3.3-70b-versatile',
    messages=chathistory
)

print(completion.choices[0].message.content)
```
